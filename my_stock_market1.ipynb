```python
#svr
```

```python
X_s = dataset.iloc[:, 1:2].values
```

```python
X_s
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_s = sc_X.fit_transform(X_s)
y_s = sc_y.fit_transform(y)
```

```python
print(X_s)
```

```python
print(y_s)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_s, y_s)
```

```python
sc_y.inverse_transform(regressor.predict(sc_X.transform([[233.42]])))
```

```python
plt.scatter(sc_X.inverse_transform(X_s), sc_y.inverse_transform(y_s), color = 'red')
plt.plot(sc_X.inverse_transform(X_s), sc_y.inverse_transform(regressor.predict(X_s)), color = 'blue')
plt.title('stock (support vector regression)')
plt.xlabel('open')
plt.ylabel('close')
plt.show()
```

```python
X_grid = np.arange(min(sc_X.inverse_transform(X_s)), max(sc_X.inverse_transform(X_s)), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(sc_X.inverse_transform(X_s), sc_y.inverse_transform(y_s), color = 'red')
plt.plot(X_grid, sc_y.inverse_transform(regressor.predict(sc_X.transform(X_grid))), color = 'blue')
plt.title('stock (support Vector Regression)')
plt.xlabel('open')
plt.ylabel('close')
plt.show()
```
